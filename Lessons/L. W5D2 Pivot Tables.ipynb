```python
import pandas as pd
import numpy as np

vehicles = pd.read_csv('vehicles.csv')
vehicles.head(50)
```

```python
vehicles.pivot_table(index=["Vehicle Class"])
```

```python
vehicles.pivot_table(index=["Vehicle Class"], columns=["Cylinders"])
```

```python
vehicles.pivot_table(index=["Vehicle Class"], columns=["Cylinders"],
                     values=["CO2 Emission Grams/Mile"], fill_value=0)
```

```python
vehicles.pivot_table(index=["Vehicle Class"], 
                     values=["Combined MPG", "Year"],
                     aggfunc=np.max)
```

```python
vehicles.pivot_table(index=["Vehicle Class"], 
                     values=["Combined MPG", "Year"],
                     aggfunc='count')
```

```python
vehicles.pivot_table(index=["Vehicle Class"], 
                     values=["Combined MPG", "CO2 Emission Grams/Mile"],
                     aggfunc=np.max)
```
