```python
import pandas as pd
import numpy as np
```

```python
matrix = [(222, 34, 23),
         (333, 31, 11),
         (444, 16, 21),
         (555, 32, 22),
         (666, 33, 27),
         (777, 35, 11)
         ]
```

```python
dfObj = pd.DataFrame(matrix, columns=list('abc'))
```

```python
dfObj
```

```python
# Apply a lambda function to each column by adding 10 to each value in each column
modDfObj = dfObj.apply(lambda x : x + 10)
 
print("Modified Dataframe by applying lambda function on each column:")
print(modDfObj)

print("but we could already do this by just adding 10 'to the dataframe'")
modDfObj2 = dfObj + 10
print(modDfObj2)

print("we can also use apply to a number of columns")
modDfObj3 = dfObj.copy()
modDfObj3[['x','y']] = modDfObj3[['x', 'y']].apply(lambda x: x+10)
print(modDfObj3)

```

```python
print(dfObj) 
print("Modified Dataframe by applying a numpy function to get sum of values in each column :")
modDfObj = dfObj.apply(np.sum) # default: axis=0
print(modDfObj)
print("Modified Dataframe by applying a numpy function to get sum of values in each row :")
modDfObj = dfObj.apply(np.sum, axis=1)
print(modDfObj)
```

```python
#applying a function to a single row or column using apply


# List of Tuples
matrix = [(22, 34, 23),
         (33, 31, 11),
         (44, 16, 21),
         (55, 32, 22),
         (66, 33, 27),
         (77, 35, 11)
         ]
 
# Create a DataFrame object
dfObj = pd.DataFrame(matrix, columns=list('xyz'), index=list('abcdef'))
```

```python
#apply on a column
#method 1:
# Apply function numpy.square() to square the value one column only i.e. with column name 'z'
modDfObj = dfObj.apply(lambda x: np.square(x) if x.name == 'z' else x)
 
print("Modified Dataframe : Squared the values in column 'z'")
modDfObj
```

```python
#method 2:
dfObj2 = dfObj.copy()
# Apply a function to one column and assign it back to the column in dataframe
dfObj2['z'] = dfObj2['z'].apply(np.square)
dfObj2
```

```python
#method 3:
dfObj3 = dfObj
# Apply a function to one column and assign it back to the column in dataframe
dfObj3['z'] = np.square(dfObj3['z'])
dfObj3
```

```python
# Apply function numpy.square() to square the values of one row only i.e. row with index name 'b'
modDfObj = dfObj.apply(lambda x: np.square(x) if x.name == 'b' else x, axis=1)
 
print("Modified Dataframe : Squared the values in row 'b'\n", modDfObj)
```

```python
# creating a new column based on values in existing columns
# List of Tuples
matrix = [(22, 34, 23),
         (33, 31, 11),
         (44, 16, 21),
         (55, 32, 22),
         (66, 33, 27),
         (77, 35, 11)
         ]
 
# Create a DataFrame object
df = pd.DataFrame(matrix, columns=list('xyz'), index=list('abcdef'))
#df
```

```python
df['u'] = df.apply(lambda row: row.x/(row.y * row.z), axis=1)
```

```python
df
```

```python
#same but only for certain rows
#lambda with if statement does not work here, so have to define a function
def myfunc(line):
    if line.x> 50:
        return (line.x/(line.y * line.z))
    
df['v'] = df.apply(lambda row: myfunc(row), axis=1)
```
