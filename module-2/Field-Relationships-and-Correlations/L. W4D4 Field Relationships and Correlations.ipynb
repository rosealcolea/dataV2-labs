```python
import numpy as np
import pandas as pd

vehicles = pd.read_csv('vehicles.csv')
# lets look if City Mileage and Highway Mileage are related, 
# that is: change in the same direction
vehicles[['City MPG', 'Highway MPG']].head(50)
```

```python
vehicles.columns
```

```python
# we can quantify this by using the corr() method
vehicles['City MPG'].corr(vehicles['Highway MPG'])
```

```python
# calling the corr() method on a dataframe gives the pairwise correlation 
# between all numeric vaiables - this is called the 'correlation matrix'
vehicles.corr()
```

```python
# 3 slightly differentways to define correlation
# 'regular' one is pearson-correlation 
# (used 99% of the time, and therefore the default)
vehicles.corr(method='pearson')
```

```python
# alternative ones are called 'spearman' and 'kendall'
# spearman correlation is useful for non-linear relationships, 
# just looks if the data moves in roughly the same direction
vehicles.corr(method='spearman')
```

```python
# kendall or "kendall's-tau" is used even more rarely
vehicles.corr(method='kendall')
```

```python
vehicles['City MPG'].corr(vehicles['Highway MPG'], method="kendall")
```
