```python
#dependent samples - before/after
import pandas as pd

blood_pressure = pd.read_csv('blood_pressure.csv')
blood_pressure.head()
```

```python
from scipy.stats import ttest_rel
# DEPENDENT samples: Here we look at the mean of the differences
# mu(after - before)
# H0: there is no difference
# H1: there is a difference
ttest_rel(blood_pressure.after, blood_pressure.before)
```

```python
from scipy.stats import ttest_1samp
ttest_1samp(blood_pressure.after - blood_pressure.before, 0)
```

```python
# independent samples, e.g., A/B tests
ab_test = pd.read_csv('ab_test.csv')
ab_test
```

```python
from scipy.stats import ttest_ind

# INDEPENDENT samples: we look at the difference of the means
# mu(after) - mu(before)
# H0: there is no difference
# H1: there is a difference

# equal variance
ttest_ind(ab_test.a, ab_test.b, equal_var=True)
```

```python
ab_test.b.mean()
```

```python
# equivalent to this 1 sample test:
ttest_1samp(ab_test.a, ab_test.b.mean())
```

```python
# unequal variance, or we don't know (Welch's test)
ttest_ind(ab_test.a, ab_test.b, equal_var=False)

```
