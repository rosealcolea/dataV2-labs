```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
vehicles = pd.read_csv('vehicles.csv')

vehicles.sort_values(by=["CO2 Emission Grams/Mile"], inplace=True)
vehicles.plot(x="CO2 Emission Grams/Mile", y=["City MPG", "Highway MPG"])
```

```python
vehicles_mean = vehicles[["Highway MPG", "City MPG", "Drivetrain"]].groupby(["Drivetrain"]).agg("mean")
vehicles_mean
```

```python
vehicles_mean.plot.bar()
```

```python
vehicles_mean.plot.barh()
```

```python
pd.plotting.scatter_matrix(vehicles, figsize=(15, 10))
```
