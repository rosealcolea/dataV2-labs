#### The data we will use in this challenge is:

```python
x = np.arange(0,100)
y = x*2
z = x**2
```

#### Plot (x, y) and (x, z).
There are 2 ways of doing this. Do in both ways.

**Hint**: Check out the nrows, ncols and index arguments of subplots. 

Also, play around with the linewidth and style. Use the ones you're most happy with.

```python
# Method 1
plt.plot(x, y, label='G(y)')
plt.plot(x, z, label='G(z)')

plt.xlabel('x')
plt.ylabel('y')

plt.title("Linear vs Non Linear Plots")
plt.legend()
plt.show()
```

#### Use plt.subplots(nrows=1, ncols=2) to create the plot.

```python
# Method 2
fig, [ax1, ax2] = plt.subplots(1,2)
ax1.plot(x,y)
ax1.set_title('G(y)')
ax1.set_xlabel('x label')
ax1.set_ylabel('y label')

ax2.plot(x,z)
ax2.set_title('G(z)')
ax2.set_xlabel('x label')
ax2.set_ylabel('y label')

plt.show()
```

#### Use your previous code but now, resize your plot.

#### Now you will have to plot both $y=x^2$ and $y=e^x$ in the same plot. Create two subplots and use a normal scale in the first one and a logarithmic scale in the second one.
**Hint**: Use `set_xscale` and `set_yscale`.

```python
# your code here
fig, [ax1, ax2] = plt.subplots(1,2)
ax1.plot(x,x**2)
ax1.set_title('G(x**2)')
ax1.set_xlabel('x label')
ax1.set_ylabel('y label')

ex = [math.exp(x) for x in x]
ax2.plot(x, ex)
ax2.set_title('G(exp(x))')
ax2.set_xlabel('x label')
ax2.set_ylabel('y label')
ax2.set_xscale("log")
ax2.set_yscale("log")

plt.show()
```

#### From the Fitbit data, we want to visually understand:

How the average number of steps change by month. Use the appropriate visualization to show the median steps by month. Is Fitbitter more active on weekend or workdays? All plots must be in the same jupyter notebook cell.

**Hints**:

* Use Months_encoded and Week_or Weekend columns.
* Use matplolib.pyplot object oriented API.
* Set your size figure to 12,4
* Explore plt.sca
* Explore plt.xticks
* Save your figures in a folder called `figures` in your repo. 

```python
# MEDIAN STEPS BY MONTH_ENCODED
# your code here

ftp1 = ft.pivot_table(columns=["Months"], aggfunc=['median'], values=['Steps']).transpose().reset_index()

fig, [ax1, ax2] = plt.subplots(1,2,figsize=(12.4,6))
a=list(ftp1.Steps)
b=list(ftp1.index)
ax1.set_title('Median Steps per Month')
ax1.set_xlabel('Months')
ax1.set_ylabel('Steps')
ax1.bar(b,a)


# MEDIAN STEPS BY WORK_OR_WEEKEND
# your code here

ftp2 = ft.pivot_table(columns=["Work_or_Weekend"], aggfunc=['median'], values=['Steps']).transpose().reset_index()

c=list(ftp2.Steps)
d=list(ftp2.index)
ax2.set_title('Median Steps per Work or Weekend')
ax2.set_xlabel('Work or Weekend')
ax2.set_ylabel('Steps')
ax2.bar(d,c)

plt.show()
```

#### Write a loop to plot 3 scatter plots of the following features:

* Minutes Lightly Active vs Steps    
* Minutes Very Active vs Steps    
* Minutes Sedentary vs Steps  

```python
# your code here
xx = ['Minutes Lightly Active', 'Minutes Very Active', 'Minutes Sedentary']
for i in xx:
    x = ft[i]
    y = ft['Steps']
    plt.scatter(y,x)
    plt.ylabel('Steps')
```

#### Plot the correlation matrix using seaborn.

```python
# your code here
tc = t.corr()

mask = np.triu(np.ones_like(tc, dtype=np.bool))

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(220, 10, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(tc, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

#### What are the most correlated features?

```python
"""
your comments here

The most correlated features are:
1. (positive) siblings and spouses abroad and parents and children aboard,
   which makes sense since families would often fall in both categories.
2. (positive) survived and fare, there seems to be a slight correlation
   between how much they paid and their chances of survival
3. (negative) the strongest correlation however, is that of the fare and the
   pclass. The higher the fare the better the class, which in this case means
   the lower the number (class 1 is better than class 3)
"""
```
