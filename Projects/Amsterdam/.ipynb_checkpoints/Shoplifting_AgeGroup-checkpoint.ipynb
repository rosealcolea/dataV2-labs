#### Importing Dataset

I'm using a dataset compiling crime complaints in New York City containing Shoplifting.  

```python
data = pd.read_csv('NYPD_Complaint_Data_Historic.csv')
```

```python
data
```

#### Data Cleaning - Null Values

I need to find nformation that shows a correlation between shoplifting and age group.

Therefore I'm going to find and delete the rows where the age group value is null.

```python
# Using this command I find that the SUSP_AGE_GROUP column has 342 null values
print(data.isnull().sum())

# I drop null values and check that it worked
missing_age = data.loc[data['SUSP_AGE_GROUP'].isnull()].index
clean_data = data.drop(missing_age)
print(clean_data.isnull().sum())

# I overrite my original file for easier access
data = clean_data
```

#### Aggregate Table

I need to compare the amount of shoplifting crimes commited per age group.

Thus I will group by AGE GROUP and count the amount of times there is an entry for each group.

```python
data_age = data.groupby(['SUSP_AGE_GROUP'])['CMPLNT_NUM'].agg(['count']).reset_index()
data_age
```

#### Exporting it to Excel

Since i need to distribute the amount of crimes per year withing each age group, I'm going to export the table to do excel and then re-import it once the calculations are done.

```python
data_excel = pd.ExcelWriter("Age_Group.xlsx", engine="xlsxwriter")
data_age.to_excel(data_excel, sheet_name = 'Sheet2')
data_excel.save()
```

```python
data2 = pd.read_excel('Age_Group.xlsx')
```

```python
data2
```

#### Drawing the Graph

Once the table is how I need it to be, I plot the values I want to showcase using matplotlib

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns

title = 'Shoplifting Crime Count Per Age Group'

fig, ax = plt.subplots(figsize=(15,8))
barchart = sns.barplot(data=data2, x='SUSP_AGE_GROUP', y='%')
plt.title(title + "\n", fontsize=16)
```

```python
fig = barchart.get_figure()
fig.savefig(title + '.png')
```
